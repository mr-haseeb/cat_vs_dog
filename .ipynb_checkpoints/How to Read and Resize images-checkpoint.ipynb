# Resize image using cv2

```python
import cv2
import matplotlib.pyplot as plt
path='PetImages/Dog/0.jpg'
img_arr=cv2.imread(path)

plt.imshow(img_arr)
```

```python
width=50
height=50
new_array=cv2.resize(img_arr,(width,height))
plt.imshow(new_array)
```
